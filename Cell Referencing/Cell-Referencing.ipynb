**Cell Referencing in Excel**

- Cell referencing is very important feature of excel as it helps us to generalize any of the operation/functions.
- these are three types of cell referencing
    - Relative
    - Absolute (Example : $A$1 for specific row and column)
    - Mixed (Example : $A1 or A$1 for specific row or column)

    Note : By default it is Relative Cell referencing.!


![alt text](<Relative & Absolute Referencing.png>)

tip - $ this sign blocks the row or column to change referance while using the operation on multiple rows and column 

Relative Referencing
- Operations on two cell or adding two whole table in excel 
    - = (cell reference 1) (operator) (cell reference 2)



Absolute Referencing
- Operations on 1 cell to multiple cell in the excel
    - = ($Column$Row) (operation) (cell references)




Marginal Referencing 




- Operations on particular rows or columns to the particular rows and columns
    - = ($column row) (operation) (column $row)
